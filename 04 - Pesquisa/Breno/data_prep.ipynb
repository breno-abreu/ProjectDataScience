# Preparação dos dados para a detecção de outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Carrega a base de dados da SiGesGuarda
df = pd.read_csv('../../02 - Analise exploratoria/Datasets/sigesguarda.csv', sep=';', encoding='latin-1')
df = df.drop(0)
df.head(10)
```

```python
# Seleciona apenas as colunas relevantes ligadas a data e hora, local e tipo de ocorrência
df = df[['OCORRENCIA_DATA', 'OCORRENCIA_ANO', 'OCORRENCIA_MES', 'OCORRENCIA_DIA_SEMANA', 'OCORRENCIA_HORA', 'ATENDIMENTO_BAIRRO_NOME', 'NATUREZA1_DESCRICAO']]
df
```

```python
# Renomeia as colunas do data frame
column_mapping = {'OCORRENCIA_DATA' : 'data', 
                  'OCORRENCIA_ANO' : 'ano',
                  'OCORRENCIA_MES' : 'mes',
                  'OCORRENCIA_DIA_SEMANA' : 'diaSemana',
                  'OCORRENCIA_HORA' : 'horario',
                  'ATENDIMENTO_BAIRRO_NOME' : 'bairro',
                  'NATUREZA1_DESCRICAO' : 'ocorrencia'}

df = df.rename(columns=column_mapping)
df
```

```python
# Converte as colunas necessárias para os tipos corretos
df['data'] = pd.to_datetime(df['data'])
df['ano'] = df['ano'].astype(int)
df['mes'] = df['mes'].astype(int)
df.info()
```

```python
# Transforma todas as colunas contendo textos para apenas letras minúsculas
df['diaSemana'] = df['diaSemana'].str.lower()
df['bairro'] = df['bairro'].str.lower()
df['ocorrencia'] = df['ocorrencia'].str.lower()
df.head(5)
```

```python
# Extrai apenas a hora da coluna 'horario' e insere o valor em uma nova coluna
df['hora'] = df['horario'].str[:2]
df['hora'] = df['hora'].astype(int)
df.head(5)
```

```python
# Adiciona uma nova coluna indicando o período de tempo da ocorrência baseado na hora
# das 0h até 6h -> madrugada
# das 6h até 12h -> manhã
# das 12h até 18h -> tarde
# das 18h até 0h -> noite

conditions = [
    (df['hora'] >= 0) & (df['hora'] < 6),
    (df['hora'] >= 6) & (df['hora'] < 12),
    (df['hora'] >= 12) & (df['hora'] < 18),
    (df['hora'] >= 18) & (df['hora'] <= 23)
]

choices = ['madrugada', 'manhã', 'tarde', 'noite']

df['periodoDia'] = np.select(conditions, choices, default='Unknown')
df.head(5)
```

```python
# Remove linhas duplicadas
df = df.drop_duplicates()
df.info()
```

```python
# Seleciona apenas 12 tipos de ocorrência. 
# Essas ocorrências são criminais, em contraste com a maior parte das ocorrências que não tem essa natureza
df = df[df['ocorrencia'].isin(['agressão física/verbal', 'crime ambiental', 'dano', 'estupro', 'furto', 'homicídio', 'maus tratos a animais', 'roubo', 'seqüestro e cárcere privado', 'substância Ilícita', 'violação de medida protetiva lei maria da penha', 'porte ilegal'])]
df.head(10)
```

```python
# Retirna linhas onde o bairro não foi informado, tem o valor 'bairro fictício' ou tem valor nulo
# Também corrige nomes de bairros não formatados corretamente
df = df[~(df['bairro'] == 'bairro nao informado')]
df = df[~(df['bairro'] == 'bairro fictício')]
df['bairro'] = df['bairro'].replace('centro ', 'centro')
df['bairro'] = df['bairro'].replace('tingüi', 'tingui')
df= df.dropna(subset=['bairro'])
df.info()
```
