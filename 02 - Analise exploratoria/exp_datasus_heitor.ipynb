```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Exporting cvs file for the DataSUS dataset
df = pd.read_csv('./Datasets/datasus_mortalidade_geral_ctba.csv', sep=';', encoding='latin-1')
df.tail(10)
```

```python
df1 = pd.read_csv('./Datasets/datasus_obitos_causa_externa_ctba.csv', sep=';', encoding='latin-1')
df1.tail(10)
```

```python
fullset = df.set_index('Ano/mês do Óbito').join(df1.set_index('Ano/mês do Óbito'))
fullset.tail(10)
```

```python
fullset.to_csv('./Datasets/datasus_obitos_fullset_ctba.csv')
```

```python
fullset.info()
```

```python
#Selecionando as colunas relevantes, estas sendo (Total de óbitos, Óbitos por agressão, indeterminados e por intervenção legal)
df_datasus = fullset[['Óbitos_p/Ocorrência', 'Agressões', 'Eventos (fatos) cuja intenção é indeterminada', 'Intervenções legais e operações de guerra']]
#Limpeza de memória
df = pd.DataFrame({})
df1 = pd.DataFrame({})
fullset = pd.DataFrame({})

#Renomeando colunas para nomes mais padronizados
df_datasus.reset_index(drop=False, inplace=True)
df_datasus = df_datasus.rename(columns = {'Ano/mês do Óbito':'MES','Óbitos_p/Ocorrência':'TOTAL_PERIODO', 'Agressões':'AGRESSAO', 'Eventos (fatos) cuja intenção é indeterminada':'INDETERMINADO',
                                          'Intervenções legais e operações de guerra':'INTERV_LEGAL'})
df_datasus.tail(10)
```

```python
#Remanejando a coluna de data para 2 colunas uma contendo o ano e a outro o mês das ocorrências
df_datasus['ANO'] = df_datasus['MES']
df_datasus['ANO'] = df_datasus['ANO'].str.replace(r'\D', '')
df_datasus['MES'] = df_datasus['MES'].str.split('/').str[0]
#df_datasus['MES'] = df_datasus['MES'].str.


#Eliminando as linhas com totais de ano
df_datasus = df_datasus[df_datasus['MES'].str.contains('1') == False]
```

```python
df_datasus = df_datasus[df_datasus['MES'].str.contains('2') == False]

#Remanejando a ordem das colunas
temp = df_datasus.columns.tolist()
temp = temp[-1:] + temp[:-1]
df_datasus = df_datasus[temp]
```

```python
#Limpeza dos dados e tipo dos dados
df_datasus.replace('-', 0, inplace=True)
df_datasus['INDETERMINADO'] = df_datasus['INDETERMINADO'].astype(int)
df_datasus['INTERV_LEGAL'] = df_datasus['INTERV_LEGAL'].astype(int)


df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Janeiro'), '01' + df_datasus.MES)
df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Fevereiro'), '02' + df_datasus.MES)
df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Março'), '03' + df_datasus.MES)
df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Abril'), '04' + df_datasus.MES)
df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Maio'), '05' + df_datasus.MES)
df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Junho'), '06' + df_datasus.MES)
df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Julho'), '07' + df_datasus.MES)
df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Agosto'), '08' + df_datasus.MES)
df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Setembro'), '09' + df_datasus.MES)
df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Outubro'), '10' + df_datasus.MES)
df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Novembro'), '11' + df_datasus.MES)
df_datasus.MES = df_datasus.MES.mask(df_datasus.MES.str.contains('Dezembro'), '12' + df_datasus.MES)

df_datasus.info()
```

```python
df_datasus.to_csv('./Datasets/datasus_obitos_ctba_clean.csv')
df_datasus.head(10)
```
