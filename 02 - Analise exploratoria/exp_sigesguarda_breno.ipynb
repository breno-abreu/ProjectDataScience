```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Exporting cvs file for the SiGesGuarda dataset
df = pd.read_csv('./Datasets/sigesguarda.csv', sep=';', encoding='latin-1')
df = df.drop(0)
df.head(10)
```

```python
df.info()
```

```python
# Get only the necessary columns
df_guarda = df[['ATENDIMENTO_BAIRRO_NOME', 'FLAG_FLAGRANTE', 'NATUREZA1_DESCRICAO', 'OCORRENCIA_DATA', 'OCORRENCIA_DIA_SEMANA', 'SECRETARIA_NOME', 'SERVICO_NOME']]

# Nullify the original dataframe to deallocate it from the memory
df = pd.DataFrame({})

df_guarda.head(10)
```

```python
# Cast non-string columns to their correct types and deletes all rows containing NaN values

sn_map = {'Sim': True, 'Não': False}
df_guarda['FLAG_FLAGRANTE'] = df_guarda['FLAG_FLAGRANTE'].map(sn_map)
df_guarda['FLAG_FLAGRANTE'] = df_guarda['FLAG_FLAGRANTE'].astype(bool)

df_guarda['OCORRENCIA_DATA'] = pd.to_datetime(df_guarda['OCORRENCIA_DATA'])

df_guarda = df_guarda.dropna()

df_guarda.info()
```

```python
# Gets number of rows in the dataset
print("Número Total de Linhas: " + str(len(df_guarda)))
```

```python
# Gets the number of neighborhoods in the dataset
n_bairros = df_guarda['ATENDIMENTO_BAIRRO_NOME'].nunique()
print(f"Quantidade de Bairros: {n_bairros}")
```
